# this is imbalance datset lets balance:

```python
from sklearn.utils import resample
```

```python
df_majority=my_data[my_data.Default_Payment==0]
df_minority=my_data[my_data.Default_Payment==1]
```

```python
df_minority_upsampled=resample(df_minority,replace=True,n_samples=23364,random_state=123)
my_data_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
my_data_upsampled.head()
```

```python
my_data_upsampled.shape
```

```python
my_data_upsampled.Default_Payment.value_counts()
```

```python
X=my_data_upsampled.drop(["Default_Payment"],1)
```

```python
y=my_data_upsampled["Default_Payment"]
```

```python
X.shape
```

```python
y.shape
```

# Importance of Features

```python
from sklearn.ensemble import ExtraTreesRegressor
selection=ExtraTreesRegressor()
selection.fit(X,y)
```

```python
selection.feature_importances_
```

```python
importance_values=(selection.feature_importances_)*100
col_names_data=X.columns
```

```python
col_names_data
```

```python
dict_data={"feature":col_names_data,"importance":importance_values}
importance_data=pd.DataFrame(dict_data)
importance_data
importance_data.sort_values(by=['importance'],ascending=False) 
```

## Train Test Split:

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

x_transform = scalar.fit_transform(X)
x_transform
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_transform,y,test_size = 0.30, random_state= 30)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Model Building

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn1=KNeighborsClassifier(n_neighbors=5)
```

```python
knn1.fit(X_train,y_train)
```

```python
y_pred1=knn1.predict(X_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred1)
```

```python
error_rate=[]
for i in range (1,25):
    knn= KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
error_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,25),error_rate)
plt.xlabel('k')
plt.ylabel("Error Rate")
plt.style.use('ggplot')
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
param_grid={'n_neighbors' : [3,5,7,9,10,11,13,15,17,19]}
```

```python
knn_cv=GridSearchCV(knn1,param_grid,cv=10)
knn_cv.fit(X,y)
```

```python
knn_cv.best_params_
```

# Building model with best Parameter

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve
y_pred_proba = knn.predict_proba(X_test)[:,1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
legend_data=roc_auc_score(y_test,y_pred_proba)
legend_data=round(legend_data,4)
legend_data
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
print('FPR')
print(fpr)
print('TPR')
print(tpr)
print('Thresholds')
print(thresholds)
```

```python
plt.style.use('ggplot')
plt.plot(fpr,tpr, label='Knn',color='orange')
plt.plot([0,1],[0,1],'k--',color='navy')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title(' ROC curve')
plt.legend(["AUC=0.8538"], loc ="lower right") 
plt.show()
```
